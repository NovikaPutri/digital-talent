```python
print ("BIG DATA (LIST NILAI) - KELAS C")
print ("NOVIKA PUTRI")
print () 

import random

#Random
#NT : Nilai Tugas
NT = []
for i in range(250):
    NT.append(random.randrange(1,100,1))

print("Nilai Tugas")
print(NT)

#Ganjil
#NG : Nilai Ganjil
print ()
print ("Nilai Ganjil")
NG = []
for n in NT:
    if n % 2 == 1: 
        NG.append(n)
        
print (NG)
print ()
    
#Nilai > 70
N70 = []
for n in NT:
    if n > 70:
        N70.append(n)
print("Jadi, nilai yang lebih besar dari 70 adalah" , len(N70))        
print ()

#Standar deviasi
total = 0
for n in NT:
    total = total + n 
average = total / len(NT)

jumlah = 0
for n in NT:
    a = (n - average)**2
    jumlah = jumlah + a
stdev = ( jumlah/(len(NT)-1))**0.5

print("Standar Deviasi =", stdev)
print ()

#Maksimum
maksimum = 0
for n in NT:
    if (n>maksimum):
        maksimum = n
print("Nilai maksimum dari list nilai ini adalah",maksimum)

# minimum
minimum = 0
for n in NT:
    if (n<minimum):
        minimum = n
print("Nilai minimum dari list nilai ini adalah",minimum)
print ()

# modus
nilaimaks=0
modus={}
for i in NT:
    count= NT.count(i)
    modus.update({i:count})

for key in modus:
    if modus[key]>nilaimaks:
        nilaimaks=modus[key]

for key in modus:
    if modus[key]==nilaimaks:
         print("Nilai modus dari list nilai ini adalah", key)
```
